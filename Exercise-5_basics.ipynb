## Q1：ライン・プロット

`numpy`モジュールの次の定数と関数を使う。
* `np.pi`は`3.141592...`（円周率）
* `np.sin()`は三角関数の正弦（サイン）を返す。
* `np.cos()`は三角関数の余弦（コサイン）を返す。

### 問(i)

`py4macro`の`xvalues()`関数を使い`0`から$4\pi$までの値（任意の個数）を`x1`に割り当てなさい。

````python
x1 = xvalues(0, 4*np.pi, 100)
````

内包標記と`np.sin()`を使い，正弦の値から構成されるリストを`ysin`に割り当てなさい。

````python
ysin = [np.sin(x) for x in x1]
````

`ysin`のグラフをプロットしなさい。

````python
plt.plot(x1, ysin)
````

### 問(ii)

同様に，`np.cos`の値のリストを`ycos`に割り当て，`np.cos`をプロットしなさい。

````python
ycos = [np.cos(x) for x in x1]
plt.plot(x1, ycos)
````

### 問(iii)

複数の図を重ねてプロットする場合は，次のコードとなる。
```
plt.plot(...)
plt.plot(...)
```

`np.sin()`と`np.cos`を一つの図に重ねてプロットしなさい。

````python
plt.plot(x1, ysin)
plt.plot(x1, ycos)
````

### 問(iv)

問(iii)の図に判例を追加するには，次のコードとなる。
```
plt.plot(..., label='ここにラベルを書く')
plt.plot(..., label='ここにラベルを書く')
plt.legend()
```

`np.sin`のプロットには「サイン」，`np.cos`のプロットには「コサイン」の凡例を表示しなさい。

````python
plt.plot(x1, ysin, label='サイン')
plt.plot(x1, ycos, label='コサイン')
plt.legend()
````

### 問(v)

問(iv)の図に以下を追加しなさい。
* タイトル「三角関数」，フォントの大きさを`30`
* 横軸のラベル「xの値」，フォントの大きさを`20`
* 縦軸のラベル「yの値」，フォントの大きさを`20`

````python
plt.plot(x1, ysin, label='サイン')
plt.plot(x1, ycos, label='コサイン')
plt.legend()
plt.title('三角関数', size=30)
plt.xlabel('xの値', size=20)
plt.ylabel('yの値', size=20)
````

### 問(vi)

問(v)の図に次の修正を付け加えなさい。
* 「サイン」
    * 線のスタイル：`--`
    * 線の幅：`4`
    * 色：赤
* 「コサイン」
    * 線のスタイル：`:`
    * 線の幅：`2`
    * 色：黒

````python
plt.plot(x1, ysin, label='サイン',
         linewidth=4, linestyle='--', color='red')
plt.plot(x1, ycos, label='コサイン',
         linewidth=4, linestyle=':', color='k')
plt.legend()
plt.title('三角関数', size=20)
plt.xlabel('xの値', size=10)
plt.ylabel('yの値', size=10)
````

### 問(vii)

問(vi)の図のサイズを`(7,3)`にして表示しなさい。

````python
plt.figure(figsize=(7,3))
plt.plot(x1, ysin, label='サイン',
         linewidth=4, linestyle='--', color='red')
plt.plot(x1, ycos, label='コサイン',
         linewidth=4, linestyle=':', color='k')
plt.legend()
plt.title('三角関数', size=30)
plt.xlabel('xの値', size=20)
plt.ylabel('yの値', size=20)
````

````python
plt.hist(x2, bins=30, color='k', alpha=0.5,
         edgecolor='white', linewidth=2)
pass
````

* 階級の数：`50`
