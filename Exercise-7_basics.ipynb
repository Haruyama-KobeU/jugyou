### 問(iv)

横軸に人口一人当たりGDPの成長率，縦軸に投資の対GDP比を使いプロットしなさい。

````python
# 人口一人当たりGDPの対数
gdp_pop_log = np.log( gdp_pop )
# 人口一人当たりGDPの成長率
gdp_pop_growth = gdp_pop_log.diff()

plt.plot(jpn['inv-gdp'],  # 横軸 
         gdp_pop_growth)  # 縦軸
plt.xlabel('投資対GDP比')
plt.ylabel('一人当たりGDP成長率')
pass
````
