## データフレームとシリーズの基本

### データフレームとシリーズについて

* エクセルのスプレッド・シートに対応するDataFrame（データ・フレーム）
* スプレッド・シートの1行もしくは1列に相当するSeries（シリーズ）

**教科書の図10.1**

行と列には，その位置を示す番号とラベルがある。
* 位置を示す番号をインデックスもしくはインデックス番号と呼ぶ。
    * 左端の`0`、`1`、`2`は行インデックス
    * 列`X`のインデックスは`0`，列`Y`のインデックスは$1$，列`Z`のインデックスは$2$
* 位置をラベルで表すこともできる。
    * `X`，`Y`，`Z`が列ラベル
    * 行ラベルは設定されていない
