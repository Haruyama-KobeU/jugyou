### 問(ii)

横軸に`gdp`，縦軸に`investment`をプロットしなさい。

````python
plt.plot('gdp', 'investment', data=jpn)
plt.xlabel('GDP')
plt.ylabel('投資')
pass
````

GDPと投資にはどのような関係がありますか。

````python
# 概ね正の関係がある。GDP=所得=産出量であり，産出量が増えると投資も増える。
# 今期の産出量が高ければ，企業は将来も高いと予想し，今期の投資が増えたと考えられる。
# しかし，バブル景気以降，正の関係が弱くなっている。
# 　　＊ 長い期間では負の相関にも見える。
# 　　＊ 短期的には（期間を短く区切ると）正の関係を伺える。
# 　　＊ 企業にとっての不確実性（将来が読みづらい）とリスクの増大によるものと解釈できる。
# また，投資は生産量だけではなく利子率にも影響されるが，その影響はこのプロットからは読み取れない。
````

### 問(vi)
