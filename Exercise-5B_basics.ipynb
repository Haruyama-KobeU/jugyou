* 次のリンクにある図では，横軸$[0,1]$と縦軸$[0,1]$の領域が$A$と$B$に分かれており，境界線は$y=x^2$で与えられる。
    * [画像のリンク](https://cloudkobeu-my.sharepoint.com/:b:/g/personal/k0102988_cloud_kobe-u_jp/Ebtuyy4HKLxIh5mDLg2LD4EBfLvrZ12B6sCiXxbVxNJIBA?e=lPaPds)
* 目的：領域`A`の面積をシミュレーションを使い近似値を求める。
* 仮定：
    * $A$と$B$からなる正方形をダーツの的と考え，ダーツ（矢）を投げるたびに必ず的にランダムにあたると仮定する。
    * ダーツが$A$に入る場合を「当たり」とする。
    * 当たりの確率は次の式で与えられる。

    $$
    A\text{の面積}
    =\text{当たりの確率}
    =\frac{A\text{の面積}}{(A+B)\text{の面積}}
    $$

* 「投げたダーツの数」が十分に大きい場合，次の近似が成立する。

    $$
    A\text{の面積}
    \approx\text{当たりの比率}
    =\frac{\text{当たりの数}}{\text{投げたダーツの数}}
    \tag{1}
    $$

＜仮定の追加＞ $y=x^2$の線は$A$と$B$のどちらの領域と考えても良いが，次を仮定する。
* 座標$(x_0,y_0)$が領域$A$にある場合は，次の不等式が成立する。

    $$
    y_0\leq x_0^2
    $$

* 座標$(x_0,y_0)$が領域$B$にある場合は，次の不等式が成立する。

    $$
    y_0>x_0^2
    $$

### 問(iv)

次のコードを書きなさい。
* 変数`N=2000`を作成する。`n`はダーツの数。
* 変数`darts = range(1, N+1)`を作成する。`1`〜`2000`の整数を準備。
* 変数`number_of_hit=0`を作成する。`number_of_hit`は「当たり」の数。
* 変数`res_lst=[]`を作成する。`res_lst`は計算結果を格納するリスト。
* 次の`for`ループに沿って、当たりの比率を`res_lst`に追加する。
    * `1`回目のループ
        * `1`つダーツを投げて、当たりの比率を計算する。
    * `2`回目のループ
        * 新たに`1`つのダーツを投げて、`1`〜`2`回目の結果を使い当たりの比率を計算する。
    * `3`回目のループ
        * 新たに`1`つのダーツを投げて、`1`〜`3`回目の結果を使い当たりの比率を計算する。
    * $\cdots\cdots$
    * `N`回目のループ
        * 新たに`1`つのダーツを投げて、`1`〜`N`回目の結果を使い当たりの比率を計算する。
* 横軸に`darts`、縦軸に`res_lst`を使い、当たりの比率の推移をプロットする。

```python
N = 2000
darts = range(1, N+1)               # ダーツの数
number_of_hit = 0                   # 当たりの数
res_lst = []                        # 結果を格納するリスト

for i in range(1, N+1):              # 1〜2000の整数を使いforループ
    
    x = random.uniform(0, 1)        # i回目のダーツのx座標
    y = random.uniform(0, 1)        # i回目のダーツのy座標
    
    if y <= x**2:                   # 当たりの場合にカウントを１増やす
        number_of_hit = number_of_hit + 1
    else:                           # 省略可
        pass                        # 省略可
        
    areaA = number_of_hit / i       # 当たりの比率を計算
    res_lst.append(areaA)           # res_lstに結果を追加

plt.plot(darts, res_lst,            # 横軸は1〜2000、当たりの比率
          linewidth=0.9,            # 線の幅
          color="black")            # 線の色
plt.title("Calculating Area A",     # タイトル
          size=12)                  # フォントの大きさ
plt.xlabel("Number of Darts", size=10)  # 横軸のラベル、フォントの大きさ
plt.ylabel("Area A", size=10)           # 縦軸のラベル、フォントの大きさ
plt.show()
```

## Q4：シミュレーション２

* 目的：円周率$\pi=3.14159\cdots$をシミュレーションを使い近似値を計算する。
* 仮定：
    * 次のリンクにあるような，一辺が`a=1`の正方形に内接する半径`r=a/2`の円を考えよう。
        * [画像のリンク](https://cloudkobeu-my.sharepoint.com/:b:/g/personal/k0102988_cloud_kobe-u_jp/EaQ5__pF0JxBgazxU7iBcjQBDGx3HFemDZqunABzWktG7g?e=S3Kmkv)
        * シミュレーションでは，円の中心を原点$(0,0)$と考える。
    * その正方形をダーツの的と考え，ダーツ（矢）を投げるたびに必ず的にランダムにあたると仮定する。
    * 円周内（円周上も含めて）にダーツが入る場合を「当たり」とする。
    * 当たりの確率は次の式で与えられる。

    $$
    \text{当たりの確率}
    =\frac{\text{円の面積}}{\text{正方形の面積}}
    =\frac{r^2\pi}{a^2}
    =\frac{(a/2)^2\pi}{a^2}
    =\frac{\pi}{2^2}
    =\frac{\pi}{4}
    $$

* 「投げたダーツの数」が十分に大きい場合，上の式を次のように近似できる。

    $$
    \text{当たりの確率}=\frac{\pi}{4}
    \approx
    \frac{\text{当たりの数}}{\text{投げたダーツの数}}
    =\text{当たりの比率}
    $$

    又は

    $$
    \pi\approx
    4\times(\text{当たりの比率})
    \tag{2}
    $$

＜参考＞
* 円の面積 $=r^2\pi$
* 原点を中心とする円の公式

    $$
    x^2+y^2=r^2
    $$
  
* 座標$(x_0,y_0)$が円の内側にある場合（当たりの場合）は，次の不等式が成立する。

    $$
    x_0^2+y_0^2\leq r^2
    $$

### 問(v)

次のコードを書きなさい。
* 変数`N=2000`を作成する。`n`はダーツの数。
* 変数`darts = range(1, N+1)`を作成する。`1`〜`2000`の整数を準備。
* 変数`number_of_hit=0`を作成する。`number_of_hit`は「当たり」の数。
* 変数`res_lst=[]`を作成する。`res_lst`は計算結果を格納するリスト。
* 次の`for`ループに沿って、当たりの比率を`res_lst`に追加する。
    * `1`回目のループ
        * `1`つダーツを投げて、この結果を使い$4\times$(当たりの比率)を計算し、`res_lst`に追加する。
    * `2`回目のループ
        * `1`つダーツを投げて、`1`〜`2`回目の結果を使い$4\times$(当たりの比率)を計算し、`res_lst`に追加する。
    * `3`回目のループ
        * `1`つダーツを投げて、`1`〜`3`回目の結果を使い$4\times$(当たりの比率)を計算し、`res_lst`に追加する。
    * $\cdots\cdots$
    * `N`回目のループ
        * `1`つダーツを投げて、`1`〜`N`回目の結果を使い$4\times$(当たりの比率)を計算し、`res_lst`に追加する。
* 横軸に`darts`、縦軸に`res_lst`を使い、当たりの比率の推移をプロットする。

```python
N = 2000
darts = range(1, N+1)               # ダーツの数
number_of_hit = 0                   # 当たりの数
res_lst = []                        # 結果を格納するリスト

for i in range(1, N+1):             # 1〜2000の整数を使いforループ
    
    x = random.uniform(-0.5, 0.5)   # i回目のダーツのx座標
    y = random.uniform(-0.5, 0.5)   # i回目のダーツのy座標
    
    if x**2 + y**2 <= (1/2)**2:     # 当たりの場合にカウントを１増やす
        number_of_hit = number_of_hit + 1
    else:                           # 省略可
        pass                        # 省略可
        
    circle_ratio = number_of_hit / i   # 当たりの比率を計算
    res_lst.append( 4*circle_ratio )   # res_lstに結果を追加

plt.plot(darts, res_lst,            # 横軸は1〜2000、当たりの比率
          linewidth=0.9,            # 線の幅
          color="black")            # 線の色
plt.title("Calculating Pi",     # タイトル
          size=12)                  # フォントの大きさ
plt.xlabel("Number of Darts", size=10)  # 横軸のラベル、フォントの大きさ
plt.ylabel("Pi", size=10)           # 縦軸のラベル、フォントの大きさ
plt.show()
```
